#### Group datasets according to treatments

```python
grouped_all = df_main.groupby('treatment_str')
grouped_first = df_main.groupby('treatment_first')
grouped_second = df_main.groupby('treatment_second')
```

#### Set order of treatment index  

```python
index_all = pd.Index(['No piece rate + Neutral', 'No piece rate + Expectations', 'No piece rate + Goal', 
                      'Low piece rate + Neutral', 'Low piece rate + Expectations', 'Low piece rate + Goal', 
                      'High piece rate + Neutral', 'High piece rate + Expectations', 'High piece rate + Goal'
                     ], dtype='object', name='Treatment')
index_first = pd.Index(['No piece rate', 'Low piece rate', 'High piece rate'], name='Treatment')
index_second = pd.Index(['Neutral', 'Expectations', 'Goal'], name='Treatment')
```

#### Table 1: Treatment table <a id='treatment'></a>

```python
treat_table = pd.crosstab(index=df_main.treatment_first, columns=[df_main.treatment_second], margins=True)
treat_table = treat_table.reindex(index_first.append(pd.Index(['All'])), axis=0)
treat_table = treat_table.reindex(index_second.append(pd.Index(['All'])), axis=1)
treat_table.to_latex(os.path.join(tables_folder, 'treat_table.tex'), bold_rows=True, float_format="%.0f")
display(treat_table)
```

#### Table 2:  Background characterisitcs of subjects <a id='balance'></a>

```python
covariates = ['age', 'female', 'education', 'mobile_device', 'latin']
pivot_index = ['treatment_first', 'treatment_second']
balance = pd.pivot_table(df_main, index=pivot_index , aggfunc=['mean', 'sem'], values=covariates, margins=True)
balance.columns.set_levels(['Mean', 'SE'], level=0, inplace=True)
balance.columns.set_levels([x.capitalize() for x in balance.columns.levels[1]], level=1, inplace=True)
balance = balance.swaplevel(axis=1).sort_index(axis=1, level=0)
balance = balance.sort_index(ascending=False, axis=0, level=0)
balance = balance.reindex(index_second.append(pd.Index([''])), axis=0, level=1)
balance.to_latex(os.path.join(tables_folder, 'balance_table.tex'), bold_rows=True, float_format="%.2f")
display(balance)
```

#### Covariate balance using OLS <a id='balance_ols'></a>

```python
ols_results = []
for variable in covariates:
    reg = '{} ~ C(treatment)'.format(variable)  
    ols = smf.ols(formula=reg, data=df_main).fit()
    ols = ols.get_robustcov_results()
    ols_results.append(ols)

auxiliary = {'N': lambda x: "{:d}".format(int(x.nobs)), 'R2': lambda x: "{:.3f}".format(x.rsquared)}
ols_balance = summary_col(ols_results, stars=True, info_dict=auxiliary, float_format='%.3f')
ols_balance.add_title('Balance')
display(ols_balance)
```

#### Figure S1: Time spent on intervention screen <a id='timer'></a>

```python
treat = index_second[1:3]
hist_params = dict(bins=20, range=(0, 60), density=True, color='0.3', alpha=0.9) 
fig, axis = plt.subplots(ncols=2, nrows=1, sharey=True, figsize=(10, 5), dpi=100)
fig.subplots_adjust(hspace=0.35, wspace=0.25)
fig.set_facecolor('w')
for i, ax in enumerate(fig.axes):
    timer = df_main[df_main.treatment_second == treat[i]]['intervention_timer']
    timer.hist(ax=ax, **hist_params)
    ax.set_title(treat[i], **font_kwargs)
    ax.grid(False)
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    ax.tick_params(axis='x', **tick_kwargs)
    ax.tick_params(axis='y', **tick_kwargs)
    ax.set_xlabel("Time spent on intervention screen in seconds", **xlabel_kwargs)
    ax.set_ylabel("Share of subjects", **ylabel_kwargs)
    ax.yaxis.set_major_formatter(FuncFormatter('{:.0%}'.format))
    mean, med = timer.mean(), timer.median() 
    ax.text(40, 0.15 , "${x}={a:.2f}$\n${b}={c:.2f}$".format(x=r"\bar{x}", a=mean, b=r"\~{x}", c=med)) 
plt.savefig(os.path.join(figures_folder, 'intervention_timer_hist.pdf'))
```

#### Check for differences in the number of empty submissions across treatments <a id='counter_empty'></a>

```python
regs = dict()
regs['reg1'] = 'counter_empty ~ low_piece_rate + high_piece_rate'
regs['reg2'] = 'counter_empty ~ expectations + goal'
regs['reg3'] = 'counter_empty ~ (low_piece_rate + high_piece_rate) * (expectations + goal)'
regs['reg4'] = regs['reg3'] + ' + age + female + education + mobile_device + latin'

ols_results = []
for k, v in regs.items():
    ols = smf.ols(formula=v, data=df_main).fit()
    ols = ols.get_robustcov_results()
    ols_results.append(ols)
    
order = [
    'low_piece_rate', 'high_piece_rate', 'expectations', 'goal',
    'low_piece_rate:expectations', 'high_piece_rate:expectations', 
    'low_piece_rate:goal', 'high_piece_rate:goal',
    'age', 'female', 'education', 'mobile_device', 'latin',
    'Intercept', 
] 

auxiliary = {'N': lambda x: "{:d}".format(int(x.nobs)), 'R2': lambda x: "{:.3f}".format(x.rsquared)}
ols_empty = summary_col(ols_results, stars=True, info_dict=auxiliary, float_format='%.3f', regressor_order=order)
ols_empty.add_title('Regression table: Number of empty submissions')
display(ols_empty)
```

#### Table 3: Treatment effects on quantity <a id='quantity'></a>

```python
regs = dict()
regs['reg1'] = 'counter_real ~ low_piece_rate + high_piece_rate'
regs['reg2'] = 'counter_real ~ expectations + goal'
regs['reg3'] = 'counter_real ~ (low_piece_rate + high_piece_rate) * (expectations + goal)'
regs['reg4'] = regs['reg3'] + ' + age + female + education + mobile_device + latin'

ols_results = []
for k, v in regs.items():
    ols = smf.ols(formula=v, data=df_main).fit()
    ols = ols.get_robustcov_results()
    ols_results.append(ols)
    
order = [
    'low_piece_rate', 'high_piece_rate', 'expectations', 'goal',
    'low_piece_rate:expectations', 'high_piece_rate:expectations', 
    'low_piece_rate:goal', 'high_piece_rate:goal',
    'age', 'female', 'education', 'mobile_device', 'latin',
    'Intercept', 
] 

auxiliary = {'N': lambda x: "{:d}".format(int(x.nobs)), 'R2': lambda x: "{:.3f}".format(x.rsquared)}
ols_quant = summary_col(ols_results, stars=True, info_dict=auxiliary, float_format='%.3f', regressor_order=order)
ols_quant.add_title('Regression table: Quantity')

path = os.path.join(tables_folder, 'ols_quantity.tex')
with open(path, "w") as latex:
    latex.write(ols_quant.as_latex())
display(ols_quant)
```

#### Reshape dataframe for panel analysis 

```python
columns_to_keep = df_main.columns[[0, 10, 12, 18, 19, 32, 33, 34, 35, 36, 37, 38]].tolist()
columns_to_melt = ['question{0}_editratio'.format(x) for x in range(1, df_main.counter.max() + 1)]

df_melted = pd.melt(df_main, 
                    id_vars=columns_to_keep,
                    value_vars = columns_to_melt,
                    var_name='question_number',
                    value_name='edit_ratio')

df_melted = df_melted.dropna(axis=0, how='any', subset=['edit_ratio'])
convert = lambda x: int(re.findall(r'\d+', x)[0])
df_melted['question_number'] = df_melted['question_number'].apply(convert)
df_melted['edit_ratio'] = df_melted['edit_ratio'].astype(float)

df_panel = df_melted.set_index(['workerid', 'question_number'])
df_panel = df_panel.sort_index()
df_panel.head()  
```

#### Table 4: Treatment effects on quality <a id='quality'></a> 

```python
regs = dict()
regs['re1'] = 'edit_ratio ~ 1 + low_piece_rate + high_piece_rate'
regs['re2'] = 'edit_ratio ~ 1 + expectations + goal'
regs['re3'] = 'edit_ratio ~ 1 + (low_piece_rate + high_piece_rate) * (expectations + goal)'
regs['re4'] = regs['re3'] + ' + age + female + education + mobile_device + latin'

re_results = []
for k, v in regs.items():
    re_result = RandomEffects.from_formula(formula=v, data=df_panel).fit(cov_type='robust')
    re_results.append(re_result)
re_qual = MyPanelModelComparison(re_results)

path = os.path.join(tables_folder, 're_quality.tex')
with open(path, "w") as latex:
    latex.write(re_qual.summary.as_latex())

display(re_qual)
```

#### Calculate the time averaged error score and merge resukts with cross-sectional dataframe

```python
avg_edit_ratio = df_panel['edit_ratio'].mean(level=0)
avg_edit_ratio = avg_edit_ratio.to_frame(name='avg_edit_ratio')
df_main = df_main.merge(avg_edit_ratio, left_on='workerid', right_on='workerid', validate='one_to_one')
```

#### Treatment effects on quality using pooled OLS <a id='quality_ols'></a> 

```python
regs = dict()
regs['re1'] = 'avg_edit_ratio ~ 1 + low_piece_rate + high_piece_rate'
regs['re2'] = 'avg_edit_ratio ~ 1 + expectations + goal'
regs['re3'] = 'avg_edit_ratio ~ 1 + (low_piece_rate + high_piece_rate) * (expectations + goal)'
regs['re4'] = regs['re3'] + ' + age + female + education + mobile_device + latin'
 
ols_results = []
for k, v in regs.items():
    ols = smf.ols(formula=v, data=df_main).fit()
    ols = ols.get_robustcov_results()
    ols_results.append(ols)

order = [
    'low_piece_rate', 'high_piece_rate', 
    'expectations', 'goal',
    'low_piece_rate:expectations', 'high_piece_rate:expectations', 
    'low_piece_rate:goal', 'high_piece_rate:goal',
    'age', 'female', 'education', 'mobile_device', 'latin',
    'Intercept', 
] 

auxiliary = {'N': lambda x: "{:d}".format(int(x.nobs)), 'R2': lambda x: "{:.3f}".format(x.rsquared)}
ols_qual = summary_col(ols_results, stars=True, info_dict=auxiliary, float_format='%.3f', regressor_order=order)
ols_qual.add_title('Regression table: Quality, OLS')

path = os.path.join(tables_folder, 'ols_quality.tex')
with open(path, "w") as latex:
    latex.write(ols_qual.as_latex())
    
display(ols_qual)
```

#### Figure 1: Scatterplots for number of fragments submitted vs. average error score  <a id='scatter_quantity_quality'></a>

```python
def get_fitted_values(df, x, y):
    y = df.loc[:,y]
    x = df.loc[:,x]
    x = sm.add_constant(x)
    ols = sm.OLS(y, x).fit()
    return ols.fittedvalues

x_var, y_var = 'counter_real', 'avg_edit_ratio'
scatter_params = dict(xlim=[0, 60], ylim=[0, 0.15], color='0.1', alpha=0.5)
with plt.style.context('classic'):
    fig, axis = plt.subplots(ncols=3, nrows=3, figsize=(12, 12), dpi=100)
    fig.set_facecolor('w')
    fig.subplots_adjust(hspace=0.3, wspace=0.3)
    for i, ax in enumerate(fig.axes):
        treatment = index_all[i]
        df_scatter = df_main[df_main.treatment_str == treatment]
        df_scatter.plot.scatter(x=x_var, y=y_var, ax=ax, **scatter_params)
        ax.plot(df_scatter[x_var], get_fitted_values(df_scatter, x_var, y_var), c='0', lw=2)
        r, p = stats.pearsonr(df_scatter[x_var], df_scatter[y_var]) 
        ax.text(25, 0.13 , r"$r={a:.3f}\:({b:.3f})$".format(a=r, b=p))
        ax.set_title(treatment, **font_kwargs)
        ax.xaxis.set_ticks_position('bottom')
        ax.yaxis.set_ticks_position('left')
        ax.tick_params(axis='x', labelsize='small', **tick_kwargs)
        ax.tick_params(axis='y', labelsize='small', **tick_kwargs)
        ax.set_xlabel("Number submitted fragments", fontsize='small', labelpad=3)
        ax.set_ylabel("Mean error score", fontsize='small', labelpad=3)     
plt.savefig(os.path.join(figures_folder, 'corr_counter_real_editratio.pdf'), bbox_inches='tight')
```

#### Table S1: Estimate slope and intercepts for quality quantity relationship across treatments <a id='ols_quantity_quality'></a>

```python
regs = dict()
regs['re0'] = 'avg_edit_ratio ~ counter_real'
regs['re1'] = 'avg_edit_ratio ~ counter_real + C(treatment)'
regs['re2'] = 'avg_edit_ratio ~ counter_real + counter_real:C(treatment)'
regs['re3'] = 'avg_edit_ratio ~ counter_real * C(treatment)'

ols_results = []
for k, v in regs.items():
    ols = smf.ols(formula=v, data=df_main).fit()
    ols = ols.get_robustcov_results()
    ols_results.append(ols)

order = ['Intercept', 'counter_real'] 
auxiliary = {'N': lambda x: "{:d}".format(int(x.nobs)), 'R2': lambda x: "{:.3f}".format(x.rsquared)}
ols_qual_quan = summary_col(ols_results, stars=True, info_dict=auxiliary, float_format='%.4f', regressor_order=order)
ols_qual_quan.add_title('Regression table: Quality vs. Quantiy')

path = os.path.join(tables_folder, 'ols_quality_quanity.tex')
with open(path, "w") as latex:
    latex.write(ols_qual_quan.as_latex())

display(ols_qual_quan)
```
